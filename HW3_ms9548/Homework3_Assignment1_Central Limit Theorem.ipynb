```python
# Matt Sloane (ms9548)
# Homework 3, Assignment 1 - Central Limit Theorem Visualization

from __future__ import print_function  # Python 3 compatibility - print function
import numpy as np
import pandas as pd
import pylab as pl
%pylab inline   

dist = np.random.randint(11,1999,100)  # Generate 100 samples sizes > 10 and < 2000

chi_dist = []
bi_dist = []
norm_dist = []
pois_dist = []
stanG_dist = []

for i in range(0, len(dist)):
    chi_dist.append(np.random.chisquare(100,dist[i]))
    bi_dist.append(np.random.binomial(1000,0.1,dist[i]))
    norm_dist.append(np.random.normal(100,10,dist[i]))
    pois_dist.append(np.random.poisson(100,dist[i]))
    stanG_dist.append(np.random.standard_gamma(100,dist[i]))  # Standard Gamma Distribution (student choice)


chi_means = []
bi_means = []
norm_means = []
pois_means = []
stanG_means = []

for j in range(0,len(dist)):
    chi_means.append(chi_dist[j].mean())
    bi_means.append(bi_dist[j].mean())
    norm_means.append(norm_dist[j].mean())
    pois_means.append(pois_dist[j].mean())
    stanG_means.append(stanG_dist[j].mean())
    
```

```python
fig = plt.figure()
plt.scatter(x = chi_means, y = dist)
plt.xlabel("Sample Mean")
plt.ylabel("Sample Size (N)")
plt.suptitle("Chi Square Mean Distribution")
capt_1 = '''Description: The scatter plot of the Chi Square Mean Distriution 
to the left shows an increasing density of concentration of values near the 
100 mean value, which is to be expected, as the distribution mean was preset 
to 100.  There are several outlier mean (relative to the distribution),
approaching values of 97 at the low end and 103 at the high end.  Overall, 
this distribution is consistent with the Central Limit Theorem, as the
number of samples generated and analyzed show increasing
clustering around the mean.'''
fig.text(1,0,capt_1)
```
